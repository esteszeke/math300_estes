```python
import sympy as symp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, n = symp.symbols('x n')
```

```python
symp.summation(x,(x,1,n-1))
```

```python
xs = np.arange(1,8)
```

```python
plt.plot(xs, xs**2, 'r', xs, (xs**2/2 - xs/2), 'b', xs, xs**1.5, 'y', xs, xs*np.log(xs), 'm')
plt.show()
```
