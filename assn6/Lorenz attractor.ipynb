```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 1000
dt = 0.01
seq = np.empty( ((n+1),3) )
seq[0,0], seq[0,1], seq[0,2] = (1,1,1)
```

```python
def lorenz(x, y, z, a=10, b=28, c=8/3):
    dx = a*(y-x) 
    dy = x*(b-z)-y
    dz = x*y-c*z
    return (dx, dy, dz)
```

```python
for i in range(n):
    dx, dy, dz = lorenz(seq[i,0], seq[i,1], seq[i,2])
    seq[i+1,0] = seq[i,0] + dt*dx # x + dt*dx
    seq[i+1,1] = seq[i,1] + dt*dy # y + dt*dy
    seq[i+1,2] = seq[i,2] + dt*dz # z + dt*dz
```

```python
plt.plot(seq[:,1], seq[:,2])
plt.show()
```
